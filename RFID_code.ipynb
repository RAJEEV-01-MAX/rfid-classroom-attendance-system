```python
#include <SPI.h>
#include <MFRC522.h>
#include <LiquidCrystal.h>

#define SS_PIN 10
#define RST_PIN 9

MFRC522 mfrc522(SS_PIN, RST_PIN);
LiquidCrystal lcd(6, 7, 5, 4, 3, 2);

const String allowedUIDs[] = {"C65E0596", "711D797B0"};

void setup() {
  Serial.begin(9600);
  SPI.begin();
  mfrc522.PCD_Init();
  lcd.begin(16, 2);
  lcd.print("Scancard");
}

bool isAllowed(String uid) {
  for (int i = 0; i < sizeof(allowedUIDs) / sizeof(allowedUIDs[0]); i++) {
    if (uid == allowedUIDs[i]) return true;
  }
  return false;
}

void loop() {
  if (!mfrc522.PICC_IsNewCardPresent()) {
    delay(150);
    return;
  }
  if (!mfrc522.PICC_ReadCardSerial()) {
    delay(150);
    return;
  }

  String uid = "";
  for (byte i = 0; i < mfrc522.uid.size; i++) {
    if (mfrc522.uid.uidByte[i] < 0x10) uid += "0";
    uid += String(mfrc522.uid.uidByte[i], HEX);
  }
  uid.toUpperCase();

  Serial.print("Scanned UID: ");
  Serial.println(uid); // Optional, for debugging

  if (isAllowed(uid)) {
    lcd.clear();
    lcd.print("Granted");
  } else {
    lcd.clear();
    lcd.print("Denied");
  }

  mfrc522.PICC_HaltA();
  mfrc522.PCD_StopCrypto1();

  delay(2000);
  lcd.clear();
  lcd.print("Scancard");
}
```
